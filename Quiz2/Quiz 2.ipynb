Step 1:
- Define a function with a parameter "TotalStats"
- If "TotalStats > 85", classification = "Top Performer"
- If "TotalStats <= 85", classification = "Non Top-Performer"
- The function returns "classification"

Step 2:
- Extract the column "TotalStats" in the dataset <br>
- Create a new variable called "player_classification" <br>
to assign each player to the correct classification

HINT: Use the ".apply()" function

````python
# Write your own code

def soccer_stats(TotalStats):
    if TotalStats > 85:
        return "Top Performer"
    if TotalStats <= 85:
        return "Non Top-Performer"
    
player_classification = fifa23_players["TotalStats"].apply(soccer_stats)
display(player_classification)
````

(d) Regression function

<font size = "3">

- Run a regression of "Overall" (left-hand side) on "Age" (right-hand side)
- What are the values of the coefficients?

[HINT: Use the commands ```smf.ols()``` and ```.fit()``` and ```.params``` <br>
from Lecture 11 ]


````python
# Write your own code

fifa23_players["x"] = fifa23_players["Age"]
fifa23_players["y"] = fifa23_players["Overall"]

b0 = 1
b1 = 2

fifa23_players["p"] = b0 + b1*fifa23_players["x"]

plt.scatter(x = fifa23_players["x"], y = fifa23_players["y"])
plt.plot(fifa23_players["x"], fifa23_players["y"], color = 'green')

plt.xlabel("Age")
plt.ylabel("Overall")
plt.show()

model = smf.ols(formula = 'y ~  x', data =fifa23_players)
results = model.fit()

b_list = results.params
print(b_list)

fifa23_players["p_estimated"] = b_list[0] + b_list[1]  * fifa23_players["x"]

#49.787345, 0.636467
````
